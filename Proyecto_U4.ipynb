# **Metodologia**

La presente investigación tiene como objetivo analizar los factores que influyen en el rendimiento académico de los estudiantes. Para ello, se utilizaron métodos estadísticos cuantitativos, específicamente las pruebas ANOVA (Análisis de Varianza) y Tukey, para evaluar las diferencias significativas entre los diferentes grupos de variables.

Diseño del estudio
El estudio se basa en un diseño no experimental, correlacional y transversal. Se recolectaron datos de una muestra de estudiantes, a incluir varias variables que pueden influir en su rendimiento académico. Las variables consideradas incluyen la participación de los padres, el acceso a recursos, las actividades extracurriculares, el nivel de motivación, entre otras. Cada una de estas variables se divide en categorías o grupos, que se comparan entre sí para evaluar su impacto en el desempeño académico de los estudiantes.

Muestra
La muestra está compuesta por estudiantes de distintas instituciones educativas. Se seleccionaron aleatoriamente participantes de acuerdo con los criterios establecidos para cada variable. En total, se analizó un conjunto de datos que incluye las evaluaciones de exámenes de los estudiantes, así como la clasificación de cada factor considerado (por ejemplo, participación parental baja, media o elevada).

Prueba ANOVA
La prueba ANOVA (Análisis de Varianza) se empleó para determinar si existen diferencias estadísticamente significativas en el rendimiento académico (puntaje de exámenes) entre los grupos definidos por las diferentes categorías de las variables independientes.

Hipótesis :
La hipótesis nula (H0) establece que no existen diferencias significativas en el rendimiento académico entre los grupos, mientras que la hipótesis alternativa (H1) postula que sí existen diferencias significativas.

Procedimiento :
Para cada variable (por ejemplo, participación parental, acceso a recursos, nivel de motivación, etc.), se realizó un ANOVA de un solo factor. Este análisis evalúa si las medias de los grupos de cada variable (bajo, medio y alto) son iguales o diferentes. Si el valor p resultante de la prueba ANOVA es menor que 0.05 (nivel de significancia), se rechaza la hipótesis nula, indicando que existen diferencias significativas entre los grupos.

F-Estadístico :
El F-estadístico es el valor que se calcula para la prueba ANOVA. Un valor alto de F sugiere que hay una gran variabilidad entre los grupos en comparación con la variabilidad dentro de los grupos, lo que indica que las diferencias entre los grupos son significativas.

Prueba Tukey
La prueba de Tukey, también conocida como comparación múltiple de medios, se utiliza después de realizar el ANOVA para identificar específicamente qué grupos son significativamente diferentes entre sí. Esta prueba permite hacer comparaciones por pares entre los grupos de cada variable independiente, ayudando a determinar entre qué niveles de una variable existen las diferencias más marcadas.

Procedimiento :
Tras obtener un resultado significativo en el ANOVA, se aplica la prueba de Tukey para realizar comparaciones por pares entre los grupos. La prueba de Tukey calcula la diferencia entre las medias de cada par de grupos, ajustando el valor p para controlar el error tipo I en las múltiples comparaciones. Si el valor p ajustado (p-adj) es menor que 0.05, se concluye que las medias de los grupos comparados son significativamente diferentes.

Interpretación :
La prueba de Tukey proporciona resultados detallados sobre cuáles son los pares de grupos que presentan diferencias significativas en sus medios. Además, se presenta un intervalo de confianza para cada comparación, lo que ofrece una mejor comprensión de la magnitud de las diferencias entre los grupos.

Análisis de los resultados
Una vez realizadas las pruebas ANOVA y Tukey, los resultados se interpretan de la siguiente manera:

Si el valor p de la prueba ANOVA es menor que 0.05, se concluye que al menos uno de los grupos es significativamente diferente de los demás. A continuación, se aplica la prueba de Tukey para identificar exactamente cuáles son los grupos que difieren entre sí.

Si el valor p ajustado de Tukey es menor que 0.05 para una comparación específica, se establece que los medios de esos dos grupos son significativamente diferentes.

Software utilizado

# Participacion parental

```python
df['participacion_parental'].unique()
```

En esta seccion se va a analizar la variable de estudio de la participacion parental, la cual tiene variables de respuesta como bajo, medio y alto.

```python
# Si todos los grupos tienen una distribución normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado'] ['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

Con esta prueba de ANOVA se puede concluir que la participacion de los padres si influye en el puntaje final o exito academico de los alumnos.

```python
# Si todos los grupos tienen una distribución normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los gráficos de Tukey para tomar una decisión (ignorando la tabla).
#.    Estrictamente hablando, tendrían que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

# Resultados

Los resultados del ANOVA (F-Estadístico = 80.46, Valor p < 0.00001) muestran diferencias significativas en el puntaje de los estudiantes según el nivel de participación parental. La prueba de Tukey confirma que la participación elevada es la más beneficiosa, seguida por los medios y finalmente la baja . Esto indica que un mayor involucramiento de los padres tiene un impacto positivo en el desempeño académico.
