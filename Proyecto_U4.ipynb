# **Metodologia**

La presente investigacin tiene como objetivo analizar los factores que influyen en el rendimiento acadmico de los estudiantes. Para ello, se utilizaron mtodos estadsticos cuantitativos, especficamente las pruebas ANOVA (Anlisis de Varianza) y Tukey, para evaluar las diferencias significativas entre los diferentes grupos de variables.

Diseo del estudio
El estudio se basa en un diseo no experimental, correlacional y transversal. Se recolectaron datos de una muestra de estudiantes, a incluir varias variables que pueden influir en su rendimiento acadmico. Las variables consideradas incluyen la participacin de los padres, el acceso a recursos, las actividades extracurriculares, el nivel de motivacin, entre otras. Cada una de estas variables se divide en categoras o grupos, que se comparan entre s para evaluar su impacto en el desempeo acadmico de los estudiantes.

Muestra
La muestra est compuesta por estudiantes de distintas instituciones educativas. Se seleccionaron aleatoriamente participantes de acuerdo con los criterios establecidos para cada variable. En total, se analiz un conjunto de datos que incluye las evaluaciones de exmenes de los estudiantes, as como la clasificacin de cada factor considerado (por ejemplo, participacin parental baja, media o elevada).

Prueba ANOVA
La prueba ANOVA (Anlisis de Varianza) se emple para determinar si existen diferencias estadsticamente significativas en el rendimiento acadmico (puntaje de exmenes) entre los grupos definidos por las diferentes categoras de las variables independientes.

Hiptesis :
La hiptesis nula (H0) establece que no existen diferencias significativas en el rendimiento acadmico entre los grupos, mientras que la hiptesis alternativa (H1) postula que s existen diferencias significativas.

Procedimiento :
Para cada variable (por ejemplo, participacin parental, acceso a recursos, nivel de motivacin, etc.), se realiz un ANOVA de un solo factor. Este anlisis evala si las medias de los grupos de cada variable (bajo, medio y alto) son iguales o diferentes. Si el valor p resultante de la prueba ANOVA es menor que 0.05 (nivel de significancia), se rechaza la hiptesis nula, indicando que existen diferencias significativas entre los grupos.

F-Estadstico :
El F-estadstico es el valor que se calcula para la prueba ANOVA. Un valor alto de F sugiere que hay una gran variabilidad entre los grupos en comparacin con la variabilidad dentro de los grupos, lo que indica que las diferencias entre los grupos son significativas.

Prueba Tukey
La prueba de Tukey, tambin conocida como comparacin mltiple de medios, se utiliza despus de realizar el ANOVA para identificar especficamente qu grupos son significativamente diferentes entre s. Esta prueba permite hacer comparaciones por pares entre los grupos de cada variable independiente, ayudando a determinar entre qu niveles de una variable existen las diferencias ms marcadas.

Procedimiento :
Tras obtener un resultado significativo en el ANOVA, se aplica la prueba de Tukey para realizar comparaciones por pares entre los grupos. La prueba de Tukey calcula la diferencia entre las medias de cada par de grupos, ajustando el valor p para controlar el error tipo I en las mltiples comparaciones. Si el valor p ajustado (p-adj) es menor que 0.05, se concluye que las medias de los grupos comparados son significativamente diferentes.

Interpretacin :
La prueba de Tukey proporciona resultados detallados sobre cules son los pares de grupos que presentan diferencias significativas en sus medios. Adems, se presenta un intervalo de confianza para cada comparacin, lo que ofrece una mejor comprensin de la magnitud de las diferencias entre los grupos.

Anlisis de los resultados
Una vez realizadas las pruebas ANOVA y Tukey, los resultados se interpretan de la siguiente manera:

Si el valor p de la prueba ANOVA es menor que 0.05, se concluye que al menos uno de los grupos es significativamente diferente de los dems. A continuacin, se aplica la prueba de Tukey para identificar exactamente cules son los grupos que difieren entre s.

Si el valor p ajustado de Tukey es menor que 0.05 para una comparacin especfica, se establece que los medios de esos dos grupos son significativamente diferentes.

Software utilizado

# Participacion parental

```python
df['participacion_parental'].unique()
```

En esta seccion se va a analizar la variable de estudio de la participacion parental, la cual tiene variables de respuesta como bajo, medio y alto.

```python
# Si todos los grupos tienen una distribucin normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['participacion_parental'] == 'bajo']['puntaje_examen'],
    df[df['participacion_parental'] == 'medio']['puntaje_examen'],
    df[df['participacion_parental'] == 'elevado'] ['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

Con esta prueba de ANOVA se puede concluir que la participacion de los padres si influye en el puntaje final o exito academico de los alumnos.

```python
# Si todos los grupos tienen una distribucin normal.
# Si no cumplen con la normalidad, toman el resultado de la prueba de Kruskal
#     y pueden usar los grficos de Tukey para tomar una decisin (ignorando la tabla).
#.    Estrictamente hablando, tendran que realizar la prueba de Dunn.
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['participacion_parental'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")
```

# Resultados

Los resultados del ANOVA (F-Estadstico = 80.46, Valor p < 0.00001) muestran diferencias significativas en el puntaje de los estudiantes segn el nivel de participacin parental. La prueba de Tukey confirma que la participacin elevada es la ms beneficiosa, seguida por los medios y finalmente la baja . Esto indica que un mayor involucramiento de los padres tiene un impacto positivo en el desempeo acadmico.
